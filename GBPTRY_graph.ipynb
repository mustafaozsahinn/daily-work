```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
now = datetime.now()
six_mths_ago = datetime.now() - relativedelta(months=6)
a_year_ago = datetime.now() - relativedelta(years=1)
```

```python
# Raw Package
import numpy as np
import pandas as pd
#Data Source
import yfinance as yf
#Data viz library
import plotly.graph_objs as go

#Download JPYAUD data
data = yf.download(tickers = 'GBPTRY=X', start= "2022-01-01", end=now)

#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

# Add titles
fig.update_layout(
    title='Great Britain Pound/Turkish Lira')

#Show
fig.show()
```
